```python
import requests 
import json
import pandas as pd
import numpy as np
from numpy import array
from datetime import datetime, timedelta
```

```python
def list_of_time():
    days_list = []
    starting_x = 1396310400
    while starting_x < 1412208000:
        days_list.append(starting_x)
        starting_x += 86400
    return days_list
```

```python
ut = list_of_time()
```

```python
def create_link():
    base = 'https://api.darksky.net/forecast/7094f90a2c5f593cd44c74c5192618dd/40.6782,-73.9442,' 
    end = '?exclude=alerts,flags'
    list_of_links = []
    for i in ut:
        list_of_links.append(base + str(i) + end)
    return list_of_links
```

```python
links_list = create_link()
```

```python
len(links_list)
```

```python
def create_weather():
    weather_data = [] 
    for i in links_list:
        x = requests.get(i)
        weather_data.append(x.content)
    return weather_data
```

```python
weather_data_list = create_weather()
```

```python
weather_data_list
```

```python
new_list = []
for i in weather_data_list:
    p = json.loads(i)
    x = {}
    for k,v in p.items():
        if k == 'daily':
            x['daily'] = v
    new_list.append(x)
new_list[1]
```

```python
cleaner_weather = []
for i in new_list:
    for k,v in i.items():
        x = v
        cleaner_weather.append(x)
cleaner_weather[0]
```

```python
cleaner_weather1 = []
for i in cleaner_weather:
    for k,v in i.items():
        x = v
        cleaner_weather1.append(x)
cleaner_weather1[0]
```

```python
cw = []
for i in cleaner_weather1:
    for x in i: 
        final_weather_list = {}
        for k,v in x.items():
            if k == 'time':
                final_weather_list['time'] = v
            if k == 'icon':
                final_weather_list['icon'] = v
            if k == 'moonPhase':
                final_weather_list['moonPhase'] = v
            if k == 'precipIntensity':
                final_weather_list['precipIntensity'] = v
            if k == 'apparentTemperatureHigh':
                final_weather_list['apparentTemperatureHigh'] = v
            if k == 'apparentTemperatureLow':
                final_weather_list['apparentTemperatureLow'] = v
            if k == 'windSpeed':
                final_weather_list['windSpeed'] = v
        cw.append(final_weather_list)
        
w = pd.DataFrame(cw)
w.head()
```

```python
w['time'] = pd.to_datetime(w['time'], unit='s')
```

```python
w['Average_Temp'] = w[['apparentTemperatureHigh', 'apparentTemperatureLow']].apply(lambda x: (x['apparentTemperatureHigh'] + x['apparentTemperatureLow'])/2, axis = 1)
```

```python
w.rename(columns={'time': 'Date', 'icon': "Forecast", 'moonPhase': 'Moon Phase', 'precipIntensity': 'Precipitation', 'windSpeed': 'Wind Speed'}, inplace=True)
```
